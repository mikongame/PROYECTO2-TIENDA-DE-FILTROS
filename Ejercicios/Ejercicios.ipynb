```python
# Bucles

# 1. Imprimir los números del 1 al 10 usando un bucle for.
numeros_1_10 = [i for i in range(1, 11)]
print(numeros_1_10)

# 2. Imprimir los números pares del 1 al 20 usando un bucle while.
n = 1
numeros_pares = []
while n <= 20:
    if n % 2 == 0:
        numeros_pares.append(n)
    n += 1
print(numeros_pares)

# 3. Usar un bucle para calcular la suma de los números del 1 al 100.
suma_1_100 = sum(range(1, 101))
print(suma_1_100)
```

```python
# Funciones

# 1. Función que toma dos números y retorna su suma.
def suma(a, b):
    return a + b

resultado_suma = suma(5, 10)
print(resultado_suma)

# 2. Función que toma un número y retorna su factorial.
def factorial(n):
    resultado = 1
    for i in range(1, n + 1):
        resultado *= i
    return resultado

resultado_factorial = factorial(5)
print(resultado_factorial)

# 3. Función que toma un número y determina si es primo.
def es_primo(n):
    if n < 2:
        return False
    for i in range(2, int(n ** 0.5) + 1):
        if n % i == 0:
            return False
    return True

resultado_primo = es_primo(7)
print(resultado_primo)

# 4. Función que recibe una lista de números y retorna la suma de ellos.
def suma_lista(lista):
    return sum(lista)

resultado_suma_lista = suma_lista([1, 2, 3, 4, 5])
print(resultado_suma_lista)

# 5. Función que recibe una cadena de texto y retorna la cadena en reversa.
def invertir_cadena(cadena):
    return cadena[::-1]

resultado_invertir_cadena = invertir_cadena("Python")
print(resultado_invertir_cadena)
```

```python
# Bucles y Funciones - Nivel Medio

# 1. Función que imprime los primeros n números de la serie de Fibonacci.
def fibonacci(n):
    a, b = 0, 1
    serie = []
    for _ in range(n):
        serie.append(a)
        a, b = b, a + b
    return serie

resultado_fibonacci = fibonacci(10)
print(resultado_fibonacci)

# 2. Función que retorna la lista de divisores de un número.
def divisores(n):
    return [i for i in range(1, n + 1) if n % i == 0]

resultado_divisores = divisores(12)
print(resultado_divisores)

# 3. Función que retorna una lista sin elementos duplicados.
def elementos_unicos(lista):
    return list(set(lista))

resultado_unicos = elementos_unicos([1, 2, 2, 3, 4, 4, 5])
print(resultado_unicos)

# 4. Función que retorna la suma de los dígitos de un número.
def suma_digitos(n):
    return sum(int(digito) for digito in str(n))

resultado_suma_digitos = suma_digitos(123)
print(resultado_suma_digitos)

# 5. Función que cuenta las vocales en una cadena.
def contar_vocales(cadena):
    return sum(1 for letra in cadena.lower() if letra in "aeiou")

resultado_contar_vocales = contar_vocales("Hola Mundo")
print(resultado_contar_vocales)

# 6. Función que retorna los primeros n elementos de una lista.
def primeros_n_elementos(lista, n):
    return lista[:n]

resultado_primeros_n = primeros_n_elementos([1, 2, 3, 4, 5], 3)
print(resultado_primeros_n)

# 7. Función que cuenta las mayúsculas y minúsculas en una cadena.
def contar_mayus_minus(cadena):
    mayusculas = sum(1 for c in cadena if c.isupper())
    minusculas = sum(1 for c in cadena if c.islower())
    return mayusculas, minusculas

resultado_mayus_minus = contar_mayus_minus("Hola Mundo")
print(resultado_mayus_minus)

# 8. Función que verifica si un número es perfecto.
def es_numero_perfecto(n):
    return sum(i for i in range(1, n) if n % i == 0) == n

resultado_perfecto = es_numero_perfecto(6)
print(resultado_perfecto)

# 9. Función que convierte un número a binario.
def a_binario(n):
    return bin(n)[2:]

resultado_binario = a_binario(10)
print(resultado_binario)

# 10. Función que encuentra la intersección de dos listas.
def interseccion_listas(lista1, lista2):
    return list(set(lista1) & set(lista2))

resultado_interseccion = interseccion_listas([1, 2, 3], [2, 3, 4])
print(resultado_interseccion)
```

```python
# 11. Función que determina si una cadena es un palíndromo.
def es_palindromo(cadena):
    cadena = cadena.replace(" ", "").lower()
    return cadena == cadena[::-1]
resultado_palindromo = es_palindromo("Giraffarig")
print(resultado_palindromo)

# 12. Función que imprime números del 1 al 50 con "FizzBuzz".
def fizzbuzz():
    resultado = []
    for i in range(1, 51):
        if i % 3 == 0 and i % 5 == 0:
            resultado.append("FizzBuzz")
        elif i % 3 == 0:
            resultado.append("Fizz")
        elif i % 5 == 0:
            resultado.append("Buzz")
        else:
            resultado.append(i)
    return resultado

resultado_fizzbuzz = fizzbuzz()
print(resultado_fizzbuzz)

# 13. Función que ordena una lista en orden ascendente.
def ordenar_lista(lista):
    return sorted(lista)

resultado_ordenar = ordenar_lista([4, 2, 9, 1, 5])
print(resultado_ordenar)

# 14. Función que retorna palabras más largas que `n` en una lista.
def palabras_mas_largas(lista_palabras, n):
    return [palabra for palabra in lista_palabras if len(palabra) > n]

resultado_palabras_largas = palabras_mas_largas(["hola", "mundo", "Python", "programación"], 5)
print(resultado_palabras_largas)

# 15. Función que calcula la serie de Fibonacci hasta un número dado.
def serie_fibonacci(n):
    a, b = 0, 1
    serie = []
    while a <= n:
        serie.append(a)
        a, b = b, a + b
    return serie

resultado_fibonacci = serie_fibonacci(20)
print(resultado_fibonacci)

# 16. Función que retorna el número más grande de una lista.
def maximo_lista(lista):
    return max(lista)

resultado_maximo = maximo_lista([10, 20, 30, 40])
print(resultado_maximo)

# 17. Función que retorna la suma de los cubos de los dígitos de un número.
def suma_cubos_digitos(n):
    return sum(int(d)**3 for d in str(n))

resultado_suma_cubos = suma_cubos_digitos(123)
print(resultado_suma_cubos) 

# 18. Función que retorna el segundo número más grande de una lista.
def segundo_mayor(lista):
    lista_unicos = list(set(lista))
    lista_unicos.sort(reverse=True)
    return lista_unicos[1] if len(lista_unicos) > 1 else None

resultado_segundo_mayor = segundo_mayor([10, 20, 30, 40])
print(resultado_segundo_mayor)

# 19. Función que verifica si dos listas tienen al menos un elemento en común.
def tienen_elemento_comun(lista1, lista2):
    return bool(set(lista1) & set(lista2))

resultado_elemento_comun = tienen_elemento_comun([1, 2, 3], [4, 5, 3])
print(resultado_elemento_comun)

# 20. Función que retorna la lista invertida.
def invertir_lista(lista):
    return lista[::-1]

resultado_lista_invertida = invertir_lista([1, 2, 3, 4, 5])
print(resultado_lista_invertida)
```

```python
# 21. Función que cuenta el número de dígitos y letras en una cadena.
def contar_digitos_letras(cadena):
    digitos = sum(1 for c in cadena if c.isdigit())
    letras = sum(1 for c in cadena if c.isalpha())
    return {"Letras": letras, "Dígitos": digitos}

resultado_contar_digitos_letras = contar_digitos_letras("Hola 1234")
print(resultado_contar_digitos_letras)

# 22. Función que retorna la suma acumulada de los números en una lista.
def suma_acumulada(lista):
    acumulada = []
    suma = 0
    for num in lista:
        suma += num
        acumulada.append(suma)
    return acumulada

resultado_suma_acumulada = suma_acumulada([1, 2, 3, 4, 5])
print(resultado_suma_acumulada)

# 23. Función que encuentra el elemento más común en una lista.
from collections import Counter
def elemento_mas_comun(lista):
    return Counter(lista).most_common(1)[0][0]

resultado_mas_comun = elemento_mas_comun([1, 2, 3, 1, 2, 1, 4])
print(resultado_mas_comun)  

# 24. Función que retorna la tabla de multiplicar de un número del 1 al 10.
def tabla_multiplicar(n):
    return {i: n * i for i in range(1, 11)}

resultado_tabla_multiplicar = tabla_multiplicar(5)
print(resultado_tabla_multiplicar)


# 25. Función que cuenta las apariciones de cada caracter en una cadena.
def contar_caracteres(cadena):
    return dict(Counter(cadena))

resultado_contar_caracteres = contar_caracteres("psicosomatizacion")
print(resultado_contar_caracteres)

# 26. Función que retorna los elementos que no están en ambas listas.
def diferencia_listas(lista1, lista2):
    return list(set(lista1) ^ set(lista2))

resultado_diferencia_listas = diferencia_listas([1, 2, 3], [3, 4, 5])
print(resultado_diferencia_listas)

# 27. Función que retorna una lista sin duplicados.
def lista_sin_duplicados(lista):
    return list(set(lista))

resultado_sin_duplicados = lista_sin_duplicados([1, 2, 2, 3, 4, 4, 5])
print(resultado_sin_duplicados)

# 28. Función que retorna la suma de los cuadrados de los números pares menores o iguales a n.
def suma_cuadrados_pares(n):
    return sum(i**2 for i in range(2, n+1, 2))

resultado_suma_cuadrados_pares = suma_cuadrados_pares(10)
print(resultado_suma_cuadrados_pares)

# 29. Función que retorna el promedio de una lista de números.
def promedio_lista(lista):
    return sum(lista) / len(lista) if lista else 0

resultado_promedio = promedio_lista([10, 20, 30, 40])
print(resultado_promedio)
```

```python
# 30. Función que retorna la cadena más larga en una lista.
def cadena_mas_larga(lista_cadenas):
    return max(lista_cadenas, key=len)

resultado_cadena_mas_larga = cadena_mas_larga(["Hola", "Python", "Programación"])
print(resultado_cadena_mas_larga)

# 31. Función que retorna los primeros n números primos.
def primeros_n_primos(n):
    primos = []
    num = 2
    while len(primos) < n:
        if all(num % i != 0 for i in range(2, int(num**0.5) + 1)):
            primos.append(num)
        num += 1
    return primos

resultado_primeros_n_primos = primeros_n_primos(5)
print(resultado_primeros_n_primos)

# 32. Función que invierte el orden de las palabras en una cadena.
def invertir_palabras(cadena):
    return " ".join(cadena.split()[::-1])

resultado_invertir_palabras = invertir_palabras("Hola mundo en Python")
print(resultado_invertir_palabras)

# 33. Función que ordena una lista de tuplas basada en el último elemento.
def ordenar_por_ultimo_elemento(lista_tuplas):
    return sorted(lista_tuplas, key=lambda x: x[-1])

resultado_ordenar_tuplas = ordenar_por_ultimo_elemento([(1, 3), (2, 2), (3, 1)])
print(resultado_ordenar_tuplas)

# 34. Función que cuenta la cantidad de vocales en una cadena.
def contar_vocales(cadena):
    return sum(1 for letra in cadena.lower() if letra in "aeiou")

resultado_contar_vocales = contar_vocales("Hola Mundo")
print(resultado_contar_vocales)

# 35. Función que determina si un número es primo.
def es_primo(n):
    if n < 2:
        return False
    for i in range(2, int(n ** 0.5) + 1):
        if n % i == 0:
            return False
    return True

resultado_es_primo = es_primo(17)
print(resultado_es_primo)

```
